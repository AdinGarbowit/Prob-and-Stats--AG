```python
current_point = 0.12345          # seed value (When generating random numbers, the seed value can be seen as the starting point. Seeds by themselves are not random, but they are necessary for generating pseudo-random numbers)
points_to_save = np.zeros(50000,)  # initialize an array of zeros (points_to_save is creating an array that consists of 50000 zeros)
X = np.zeros(50000)

for i in range(X.shape[0]): #The .shape functions returns the shape of an array. An array's shape is its size. An array such as (-1,0,1,2) has a shape of 4. The array above, which has 50000 zeroes, has a shape of 50000. This means that the process in this line of code is being repeated 50000 times
    current_point = 3*current_point % 1      # update x and store the value in X (this line is redefining the seed point as taking the remainder of 3 times it)
    points_to_save[i] = current_point #this line updates points_to_save as the redefined version of current_point. The current point will be multipled by 3 and taken the remainder of for 50000 iterations, replacing each zero in the array with the newly generated number. This creates pseudo-random numbers.

plt.plot(points_to_save, '.')     # plot the values in X (This provides a visual element to the pseudo-random number. The random numbers are between 0 and 1)
```

```python
plt.hist(points_to_save); #Instead printing it in dot form like the previous cell, this prints points_to_save as a histogram
```

```python
def coin():
    return 2*(np.random.random() > 0.5) - 1 #In this line, the true or false that comes from the random number being greater than or less than 0.5 can be seen as having a value. True equals 1 and false equals 0. This means that if true, the result is +1, and if false, the result is -1
```

```python
coin() #This runs the above function, either printing 1 or -1
```

```python
x = 0
x + coin() #This also prints either 1 or -1. Since x = 0, the result is either 0 +1 or 0 -1
```

```python
x = 0

for i in range(3):
    x = x + coin() #This cell does the same thing as the above cell, except that the process is repeated 3 times. This can result in numbers different from -1 or 1. Because of normal distribution, though, it is likely that the number will be close to -1, 1, or 0
```

```python
x = 0

for i in range(100):
    x = x + coin() #This repeats the process 100 times, and makes the principles of normal distribution more clear. It is statistically more likely for the number to fall around 1 and -1 because there is an equal chance of the generated x being -1 or 1.
```

```python
M = 10000
N = 100

X = np.zeros(M,) #This is creating a an array with zeros the size of M (in this case 10000)

for j in range(M): #This repeats the process based on the size of M (M = 10000 means that the process will repeat 10000 times)

    x = 0

    for i in range(N): #This repeats the process N times (N = 100)
        x = x + coin() #This code is the same as the above cell which simulates flipping a coin and then adding that value of the coin flip to the next coin flip (repeating N times)

    X[j] = x
```

```python
plt.hist(X); #This plots the generated numbers as a histogram. The histogram is in a shape very close to the bell curve
```

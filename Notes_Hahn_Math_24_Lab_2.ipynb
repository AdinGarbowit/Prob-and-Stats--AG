#Code for Printing Array of Pascal's Triangle

```python
P = np.zeros((12,18)) #When an ordered pair is written after .zeros, an array of zeros with those dimensions are created ((x,y) creates x rows and y columns)

P[0,5]=1 #defines the position 0x5 as 1

for i in range(1,P.shape[0]):

    for j in range(1,P.shape[1]-1):

        P[i,j] = P[i-1,j-1]+P[i-1,j] #Adds the latest, nonzero element of a row to the values that proceed it. Performs this action until each row's last nonzero number is 1

print(P[:,5:]) #prints the result of the above code (an array that is Pascal's triangle)
```

# Statistical Functions

```python
def mean(x):
    return np.sum(x)/x.shape[0]
```

```python
def var(x):
    return mean((x - mean(x))**2)
```

```python
def std(x):
    return np.sqrt(var(x))
```

```python
def median(x): #function for calculating the median of data
    n = len(x) #len(x) returns the number of items in a container
    sorted_x = np.sort(x) #.sort is used to put the data into ascending order
    mid = n // 2 #the number of items in the container are being floor divided by 2
    if n % 2 == 0: #if the result of the above operation is divisble by 2 (even), proceed
        return (sorted_x[mid - 1] + sorted_x[mid]) / 2
    else: #if the number is not divisble by 2 (odd)
        return sorted_x[mid]
```

```python
def mode(x): #mode is the most frequently occuring value or item in an array
    values, counts = np.unique(x, return_counts=True) #takes the unique values in x and counts how many times that unique value occurs in the set
    max_count_index = np.argmax(counts) #finds the unique value with the highest frequency
    return values[max_count_index] #returns the value that appeared most in the set
```

```python
def range(x):
    return np.max(x) - np.min(x) #.max returns the max value and .min returns the minimum value. The range of an array is calcuated by subtracting the max by the min
```

```python
data = np.array([1, 2, 2, 3, 4, 5, 5, 5, 6]) #data for testing statistic functions

# Testing the functions (Calls the functions all defined above using the above data)
mean_value = mean(data)
var_value = var(data)
std_value = std(data)
median_value = median(data)
mode_value = mode(data)
range_value = range(data)

mean_value, var_value, std_value, median_value, mode_value, range_value #sepearting the functions by commas will seperate the output by commas
```

# Approximating the Value of Pi from Random Numbers

```python
N = 10
```

```python
points = -1 + 2*np.random.random((N,2)) #generates a random number between 0 and 1, adds -1 to it, and puts it in a 10x2 array
```

```python
points #prints the array of points
```

```python
plt.plot(points[:,0],points[:,1],'.') #plots the points as dots. Points from the first column are the x component and points from the second column are y component

plt.gca().set_aspect(1) #sets the aspect ratio of the figure
```

```python
inside_circle  = points[:,0]**2 + points[:,1]**2  <=  1 #this checks if the the first column squared plus the second column squared is less than or equal to 1. Defined as inside_circle
outside_circle = points[:,0]**2 + points[:,1]**2  > 1 #this checks if the second column squared plus the second column squared is greater than 1. Defined as outside_circle

plt.plot(points[inside_circle,0],points[inside_circle,1],'g.') #plots the points defined as the inside circle as green dots and plots the points defined as the outside circle as rad dots
plt.plot(points[outside_circle,0],points[outside_circle,1],'r.')


plt.gca().set_aspect(1)
```

```python
np.sum(inside_circle),np.sum(outside_circle) #takes the sum of the data for the inside circle and the data for the sum of the outside circle and prints them as an ordered pair
```

```python
total_area = 4 #total_area is being defined as 4
```

```python
fraction_inside = np.sum(inside_circle)/N #the fraction of the inside is the sum of the inside circle points divded by N (N = 10, the number of rows in the data array)
```

```python
fraction_inside*total_area #the fraction inside as defined above is being multiplied by the total area, which is defined as 4. This is a very rough estimation of the value of pi. As N becomes greater, the accuracy will increase
```

```python
#Pi estimation combined together
N = 100000000 #N is much larger then the N used in the code above. This means pi will be much more accurate
points = -1 + 2*np.random.random((N,2))
inside_circle  = points[:,0]**2 + points[:,1]**2  <=  1
fraction_inside = np.sum(inside_circle)/N
fraction_inside*4
#The result of this code is accurated to 3.141, afterwards, it diverges
```

# Complete Code for Estimating π using Monte Carlo Simulation

```python
# Number of random points to generate
num_points = 10000 #same as the N used in the above code

# Generating random points
x_points = np.random.uniform(-1, 1, num_points)
y_points = np.random.uniform(-1, 1, num_points)

# Calculating the number of points inside the quarter circle
points_inside = np.sqrt(x_points**2 + y_points**2) <= 1
num_inside = np.sum(points_inside)

# Estimating π
pi_estimate = 4 * num_inside / num_points

# Plotting the points and the quarter circle
plt.figure(figsize=(6, 6))
plt.scatter(x_points[points_inside], y_points[points_inside], color='green', label='Inside') #plots the inside points as green and labels them inside
plt.scatter(x_points[~points_inside], y_points[~points_inside], color='red', label='Outside') #plots the outside points as red and labels them outside
circle = plt.Circle((0, 0), 1, color='blue', fill=False) #creates a circle that can be used to compare to the estimation of the circle. Can be used to visualize how accurate the estimation was. fill = false means that the circle won't fill and obscure the plot
plt.gca().add_artist(circle)
plt.title('Estimating π using Monte Carlo Simulation') #.title is used to create title of plot
plt.xlabel('x') #.xlabel creates the x label
plt.ylabel('y') #.ylabel creates the y label
plt.legend() #.legend creates a legend for the colors in the plot
plt.axis('equal') #.axis('equal') creates an equal aspect ratio
plt.show() #displays the plot

pi_estimate
```

# Approximating the Value of e from Random Numbers

```python
X = np.random.random((1000000,10)) #creates a 1000000x10 array of random numbers
```

```python
X #prints the array X
```

```python
Y = np.cumsum(X,1) #.cumsum is cumulative sum. It adds every item in an array by the values previous to it. The one indicates that it is starting in axis 1 of the array
```

```python
Y
```

```python
Z = np.argmax(Y > 1,1) + 1 #This searches for where the cumulative sum calculated above is greater than 1. If true, it adds 1
```

```python
Z #prints the array values of Z
```

```python
np.mean(Z) #prints the mean value of Z (estimation of e)
```

```python
np.mean(np.argmax(np.cumsum(np.random.random((10000000,10)),1) > 1,1) + 1) #this combines all of the above steps into one function estimating the value of e
```
